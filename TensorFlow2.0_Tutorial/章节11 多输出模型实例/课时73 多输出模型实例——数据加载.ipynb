# 1. 数据读取与处理

```python
data_dir = 'D:/BaiduNetdiskDownload/日月光华-tensorflow资料/数据集/multi-output-classification多输出模型数据集/multi-output-classification/dataset'
data_root = pathlib.Path(data_dir)
all_images_path = list(data_root.glob('*/*'))
images_count = len(all_images_path)
images_count
```

```python
# 对所有的图片地址做一个乱序
all_images_path = [str(path) for path in all_images_path]
random.shuffle(all_images_path)
# all_images_path
```

```python
# 从图片地址提取label名称
labels_name = sorted(item.name for item in data_root.glob('*/') if item.is_dir())
# 由于提取到的label是包含两个标签的，所以需要将其分开
color_labels = set(name.split('_')[0] for name in labels_name)
item_labels = set(name.split('_')[1] for name in labels_name)
```

```python
color_labels_to_index = dict((name, index) for index, name in enumerate(color_labels))
item_labels_to_index = dict((name, index) for index, name in enumerate(item_labels))
```

```python
all_image_labels = [pathlib.Path(path).parent.name for path in all_images_path]
color_labels = [color_labels_to_index[label.split('_')[0]] for label in all_image_labels]
item_labels = [item_labels_to_index[label.split('_')[1]] for label in all_image_labels]
```

# 2. 加载和格式化图像

```python
img_path = all_images_path[0]
img_path
```

```python
img_raw = tf.io.read_file(img_path)
img_tensor = tf.image.decode_image(img_raw)
img_tensor.shape, img_tensor.dtype
```

```python
img_tensor = tf.cast(img_tensor, tf.float32)
img_tensor = tf.image.resize(img_tensor, [224, 224])
img_final = img_tensor / 255
```

```python
def load_and_preprocess_image(path):
    image = tf.io.read_file(path)
    image = tf.image.decode_jpeg(image, channels=3)
    image = tf.image.resize(image, size=[224, 224])
    image = tf.cast(image, tf.float32)
    image = image / 255
    image = 2*image - 1
    return image
```

```python
path_dataset = tf.data.Dataset.from_tensor_slices(all_images_path)
AUTOTUNE = tf.data.experimental.AUTOTUNE 
images_dataset = path_dataset.map(load_and_preprocess_image, num_parallel_calls=AUTOTUNE)
labels_dataset = tf.data.Dataset.from_tensor_slices((color_labels, item_labels))

image_labels_dataset = tf.data.Dataset.zip((images_dataset, labels_dataset))
image_labels_dataset
```

```python
test_count = int(images_count * 0.2)
train_count = images_count - test_count
train_data = image_labels_dataset.skip(test_count)
test_data = image_labels_dataset.take(test_count)
BATCH_SIZE = 32
```

```python
train_data = train_data.shuffle(buffer_size=train_count).repeat(-1).batch(batch_size=BATCH_SIZE)
train_data = train_data.prefetch(buffer_size=AUTOTUNE)
train_data
```

```python
test_data = test_data.batch(batch_size=BATCH_SIZE)
```

# 3. 建立多输出模型

```python
inputs = tf.keras.Input(shape=(224, 224, 3))
# 这里仅仅是使用mobile-net的架构，而不是其预训练完毕的模型
mobile_net = tf.keras.applications.MobileNetV2(input_shape=(224, 224, 3), 
                                               include_top=False)
```

```python
# 建立函数式API以方便多输出模型的搭建
x = mobile_net(inputs=inputs)
# 在传入到下面两个不同的分类器之前，mobile-net输出的结果x是一个四维的张量，所以将其进行GlobalAveragePooling2D操作
x = tf.keras.layers.GlobalAveragePooling2D()(x)
x1 = tf.keras.layers.Dense(units=1024, activation='relu')(x)
output_color = tf.keras.layers.Dense(units=3, activation='softmax', name='output_color')(x1)
```

```python
x2 = tf.keras.layers.Dense(units=1024, activation='relu')(x)
output_item = tf.keras.layers.Dense(units=4, activation='softmax', name='output_item')(x2)
```

```python
model = tf.keras.Model(inputs=inputs, outputs=[output_color, output_item])
model.summary()
```

```python
model.compile(optimizer='adam', 
              loss={'output_color':'sparse_categorical_crossentropy',
                    'output_item':'sparse_categorical_crossentropy'},
              metrics=['acc'])
```

```python
train_steps = train_count // BATCH_SIZE
test_steps = test_count // BATCH_SIZE
```

```python
model.fit(train_data,
          steps_per_epoch=train_steps,
          validation_data=test_data,
          validation_steps=test_steps,
          epochs=3)
```
